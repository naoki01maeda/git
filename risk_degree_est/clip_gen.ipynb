```python
import pickle
import os
path = "C:/Users/naoki/anaconda_work/Faster-R-CNN-pytorch-main/drama/drama-datasets/"
files = os.listdir(path)
img_list = []
for i in range(len(files)):
  with open(path + '{0:06}.pkl'.format(i), 'rb') as web:
    img_list.append(pickle.load(web))
```

```python
import glob
import shutil
from PIL import Image, ImageSequence
from pathlib import Path

DEBUG_MODE = True

def get_frames(path):
    '''
    '''
    im = Image.open(path)
    return (frame.copy() for frame in ImageSequence.Iterator(im))

def write_frames(frames, name_original, destination):
    '''
    '''
    path = Path(name_original)

    stem = path.stem
    extension = path.suffix

    # 
    dir_dest = Path(destination)
    if not dir_dest.is_dir():
        dir_dest.mkdir(0o700)
        if DEBUG_MODE:
            print('Destionation directory is created: "{}".'.format(destination))

    for i, f in enumerate(frames):
        name = '{}/{}{}'.format(destination, i + 1, ".jpg")
        if f.mode != "RGB":
            f = f.convert("RGB")
        f.save(name)
        if DEBUG_MODE:
            print('A frame is saved as "{}".'.format(name))

for i in range(len(files)):
    save_path = 'C:/Users/naoki/anaconda_work/Faster-R-CNN-pytorch-main/data/ori_data/drama_clip/{}'.format(i)
    if not os.path.isdir(save_path):
        os.makedirs(save_path)
    clip_path = img_list[i]["img_path"].split("/")[7:-1]
    path = "G:/drama_data/combined/" + "/".join(clip_path)
    gif_files = glob.glob(path + "/*.gif")
    shutil.copy(gif_files[0], save_path)
    
    IMAGE_PATH = path + "/movie.gif"
    DESTINATION = save_path
    frames = get_frames(IMAGE_PATH)
    write_frames(frames, IMAGE_PATH, DESTINATION)
```
